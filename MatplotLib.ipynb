```python
data={'year':[1920,1930,9140,1950,1960,1970,1980,1990,2000,2010,2020],
      'Exchange Rate': [65,69,71,64,62,59,72,71,75,78,81]}
df=pd.DataFrame(data, columns=['Year', 'Exchange Rate'])


# creting barrage chart
df.plot(x ='Year',y = 'Exchange Rate', kind = 'barh')
#plt.show()
```

```python
data={'year':[1920,1930,9140,1950,1960,1970,1980,1990,2000,2010,2020],
      'Exchange Rate': [65,69,71,64,62,59,72,71,75,78,81]}
df=pd.DataFrame(data, columns=['Year', 'Exchange Rate'])


# creting scatter chart
df.plot(x ='Year',y = 'Exchange Rate', kind = 'scatter')
#plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Year'], df['Exchange Rate'], color='steelblue')
plt.xlabel('Year')
plt.ylabel('Exchange Rate')
plt.title('Exchange Rate Over Time')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
